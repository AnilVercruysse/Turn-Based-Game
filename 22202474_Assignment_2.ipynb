```python
import random


class Creature:
    def __init__(self, name, max_hp=10):
        self.name = name
        self.max_hp = max_hp
        self.hp = max_hp
        self.abilities = {"Attack": 1, "Defense": 5, "Speed": 5}

    def check_life(self):
        if self.hp <= 0:
            self.hp = 0
            print(self.name + ' has fainted...')
        return self.hp

    def attack(self, target):
        roll = random.randint(1, 20)
        if roll < target.abilities["Defense"] + target.abilities["Speed"]:
            print(self.name + ' Attacks ' + target.name + '\n' + 'Attack missed...')
            return False

        x = (random.randint(1, 4) + self.abilities["Attack"])
        target.hp -= x
        print(self.name + ' Attacks ' + target.name + '\n' + 'Attack hits for ' + str(x) + ' damage!')
        return True

    def turn(self, round_num, target):
        if target is None:
            return
        if self.attack(target):
            if target.check_life() == 0:
                return True
        return False


class Fighter(Creature):
    def __init__(self, name, max_hp=50):
        super().__init__(name, max_hp=max_hp)
        self.abilities = {"Attack": 5, "Defense": 10, "Speed": 3}
        self.shield_active = False

    def shield_up(self):
        if not self.shield_active:
            self.abilities["Attack"] -= 5
            self.abilities["Defense"] += 5
            self.shield_active = True

    def shield_down(self):
        if self.shield_active:
            self.abilities["Attack"] += 5
            self.abilities["Defense"] -= 5
            self.shield_active = False

    def turn(self, round_num, target):
        a = [1, 5, 9, 13, 17]
        if round_num in a:
            print(self.name, 'takes a defensive stance.')
            self.shield_up()
        elif round_num % 4 == 0:
            print(self.name, "'s stance returns to normal.")
            self.shield_down()

        if self.attack(target):
            if target.check_life() == 0:
                return True

        return False


class Archer(Creature):
    def __init__(self, name, max_hp=30):
        super().__init__(name, max_hp=30)
        self.max_hp = max_hp
        self.hp = max_hp
        self.abilities = {"Attack": 7, "Defense": 8, "Speed": 8}
        self.sneak = False

    def sneak_attack(self, target):
        roll = [random.randint(1, 20), random.randint(1, 20)]
        roll.sort()
        attack_roll = roll[1]

        if self.abilities["Speed"] > target.abilities["Speed"]:
            attack_roll += self.abilities["Speed"] - target.abilities["Speed"]

        if not self.sneak:
            self.abilities["Attack"] += 3
            self.abilities["Defense"] -= 3
            self.sneak = True

        x = (random.randint(1, 8) + self.abilities["Attack"])
        target.hp -= x
        print(self.name + ' sneak attacks ' + target.name + '\n' + 'Sneak attack hits for ' + str(x) + ' damage!')
        return True

    def attack(self, target):
        if self.sneak:
            self.abilities["Attack"] -= 3
            self.abilities["Defense"] += 3

            self.sneak = False

        return super().attack(target)

    def turn(self, round_num, target):
        b = [1, 5, 9, 13, 17]
        if round_num in b:
            if self.attack(target):
                if target.check_life() == 0:
                    return True
        else:
            if self.sneak_attack(target):
                if target.check_life() == 0:
                    return True
        return False


class Goblin(Creature):
    def __init__(self, name, max_hp=15):
        super().__init__(name, max_hp=max_hp)
        self.abilities = {"Attack": 4, "Defense": 6, "Speed": 6}


class Orc(Creature):
    def __init__(self, name, max_hp=50):
        super().__init__(name, max_hp=max_hp)
        self.abilities = {"Attack": 10, "Defense": 6, "Speed": 2}
        self.is_heavy_attacking = False

    def heavy_attack(self, target):
        if not self.is_heavy_attacking:
            self.abilities["Attack"] += 5
            self.abilities["Defense"] -= 3
            self.is_heavy_attacking = True

        return super().attack(target)

    def attack(self, target):
        if self.is_heavy_attacking:
            self.abilities["Attack"] -= 5
            self.abilities["Defense"] += 3
            self.is_heavy_attacking = False

        return super().attack(target)

    def turn(self, round_num, target):
        if round_num % 4 == 1 or round_num % 4 == 2 or round_num % 4 == 3:
            if self.attack(target):
                if target.check_life() == 0:
                    return True
        else:
            print(self.name, 'is in rage.', self.name, 'uses heavy attack on', target.name)
            if self.heavy_attack(target):
                if target.check_life() == 0:
                    return True
        return False


class OrcGeneral(Orc, Fighter):
    def __init__(self, name, max_hp=100):
        super().__init__(name, max_hp=max_hp)
        self.abilities = {"Attack": 10, "Defense": 6, "Speed": 2}
        self.is_heavy_attacking = False

    def turn(self, round_num, target):
        if round_num in [1, 5, 9, 13, 17]:
            if self.attack(target):
                if target.check_life() == 0:
                    return True
            print(self.name + ' takes a defensive stance.')
            self.shield_up()
        elif round_num in [2, 6, 10, 14, 18]:
            if self.attack(target):
                if target.check_life() == 0:
                    return True
        elif round_num in [3, 7, 11, 15, 19]:
            print(self.name + "'s stance returns to normal")
            self.shield_down()
            if self.attack(target):
                if target.check_life() == 0:
                    return True
        elif round_num in [4, 8, 12, 16, 20]:
            print(self.name, 'is in rage.', self.name, 'uses heavy attack on', target.name)
            if self.heavy_attack(target):
                if target.check_life() == 0:
                    return True

        if target.check_life() > 0:
            return False
        else:
            return True


class GoblinKing(Goblin, Archer):
    def __init__(self, name):
        super().__init__(name, max_hp=50)
        self.abilities = {"Attack": 4, "Defense": 6, "Speed": 6}
        self.sneak = False

    def turn(self, round_num, target):
        b = [1, 5, 9, 13, 17]
        if round_num in b:
            if self.attack(target):
                if target.check_life() == 0:
                    return True
        else:
            if self.sneak_attack(target):
                if target.check_life() == 0:
                    return True
        return False


class Wizard(Creature):
    def __init__(self, name, max_hp=20):
        super().__init__(name, max_hp=max_hp)
        self.name = name
        self.max_hp = max_hp
        self.hp = max_hp
        self.abilities = {"Attack": 3, "Defense": 5, "Speed": 5, "Arcana": 10}
        self.mana = 100

    def attack(self, target):
        if self.mana + 20 > 100:
            self.mana = 100
            print('Mana is 100')
        else:
            self.mana += 20
            print('Mana: +20!')
        roll = random.randint(1, 20)
        if roll < target.abilities["Defense"] + target.abilities["Speed"]:
            print(self.name + ' Attacks ' + target.name + '\n' + 'Attack missed...')
            return False

        x = (random.randint(1, 4) + self.abilities["Attack"])
        target.hp -= x
        print(self.name + ' Attacks ' + target.name + '\n' + 'Attack hits for ' + str(x) + ' damage!')
        return True

    def recharge(self):
        if self.mana <= 70:

            self.mana += 30
            print(self.name + " channels magical energy...\nMana: +30!")
        else:

            self.mana = 100
            print(self.name + " channels magical energy...\nMana is now 100")

    def fire_bolt(self, target):
        roll = random.randint(1, 20)
        arcana = self.abilities["Arcana"]
        if roll < target.abilities["Defense"] + target.abilities["Speed"]:
            print(self.name + ' Attacks ' + target.name + '\n' + 'Fire bolt missed...')
            return False

        damage = random.randint(1, arcana)
        target.hp -= damage
        print(self.name + ' fires a fire bolt at ' + target.name + '\n' + 'Fire bolt hits for ' + str(damage) + ' damage!\n10 Mana restored')
        self.mana += 10
        if self.mana > 100:
            self.mana = 100

        return True

    def heal(self, target):
        if self.mana < 20:
            print("The effect failed because the mana is below 20.")
            return False

        arcana = self.abilities["Arcana"]
        heal_amount = random.randint(0, 8) + (arcana // 2)
        self.mana -= 20

        heal_amount = min(heal_amount, target.max_hp - target.hp)
        target.hp += heal_amount
        print(self.name + ' Heals ' + target.name + '\n' + 'Heals for ' + str(heal_amount) + ' points!\nMana: -20')

        return True

    def mass_heal(self, allies):
        if self.mana < 30:
            print("The effect failed because the mana is below 30.")
            return False
        print(self.name, 'uses mass Heal.\nMana: -30.')
        arcana = self.abilities["Arcana"]
        heal_amount = random.randint(0, 10) + arcana
        self.mana -= 30

        for ally in allies:
            amount_to_heal = min(heal_amount, ally.max_hp - ally.hp)


            ally.hp += amount_to_heal
            print(self.name + ' Heals ' + ally.name + '\n' + 'Heals for ' + str(amount_to_heal) + ' points!')

        return True

    def fire_storm(self, enemies):
        if self.mana < 50:
            print("The effect failed because the mana is below 50.")
            return False
        print(self.name, 'uses firestorm\nMana: - 50')
        self.mana -= 50
        arcana = self.abilities["Arcana"]
        speed = self.abilities["Speed"]
        fire_storm_roll = random.randint(1, 20) + speed


        for enemy in enemies:
            if fire_storm_roll >= arcana:
                damage = (random.randint(5, 20) + arcana) // 2
            else:
                damage = (random.randint(5, 20) + arcana)
            enemy.hp = min(enemy.hp - damage, enemy.max_hp)

            print(f'{enemy.name} takes {damage} fire damage!')

        return True


class Battle:
    def __init__(self):
        self.enemies = [GoblinKing('Goblin King'), OrcGeneral('Orc General'), Goblin('Goblin'), Orc('Orc')]

        self.allies = [Fighter('Aragorn'), Archer('Legolas'), Archer('Kamelija')]

        self.player = Wizard('Gandalf')

    def auto_select(self, target_list):
        valid_targets = [target for target in target_list if target.hp > 0]

        if not valid_targets:
            return target_list[0]

        return random.choice(valid_targets)

    def select_target(self, target_list):
        print("Select target:")
        for i, target in enumerate(target_list):
            print("{}: {}, HP: {}/{}".format(i + 1, target.name, target.hp, target.max_hp))


        while True:
            choice = input("Enter choice: ")

            try:
                choice = int(choice)
                if 1 <= choice <= len(target_list):
                    return target_list[choice - 1]
            except ValueError:
                print("Invalid input. Please enter a number between 1 and {}.".format(len(target_list)))

    def start(self):
        round_num = 1

        # Add the player to the list of creatures
        creatures = sorted(self.enemies + self.allies + [self.player], key=lambda c: c.abilities["Speed"], reverse=True)

        print("=" * 30)
        print("THE BATTLE BEGINS!")

        while True:
            print("=" * 30)
            print(f"Round: {round_num}")
            print("=" * 30)
            for creature in creatures:
                if creature == self.player:
                    result = self.player_turn()

                    if result == "Quit":
                        print("=" * 30)
                        print("Player has quit.")
                        print("=" * 30)
                        break
                    elif result == "Defeat":
                        print("=" * 30)
                        print("Player has been defeated.")
                        print("=" * 30)
                        break
                else:
                    if creature.hp > 0:
                        target = self.auto_select(self.enemies if creature in self.allies else self.allies)

                        creature.turn(round_num, target)

            round_num += 1

            if all(c.hp <= 0 for c in self.enemies):
                print("=" * 30)
                print("GAME ENDS - All enemies are defeated. Allies win!")
                print("=" * 30)
                break

            if all(c.hp <= 0 for c in self.allies):
                print("=" * 30)
                print("GAME ENDS - All allies are defeated. Enemies win!")
                print("=" * 30)
                break

            # Check if the player is defeated.
            if self.player.hp <= 0:
                print("=" * 30)
                print("GAME ENDS - Player defeated. Enemies win!")
                print("=" * 30)
                break

            print("=" * 30)
            # Display a message to indicate that the round ends.
            print("Round ends.")

    def player_turn(self):
# only allow player to play if the enemies / allies are alive 
        if not (all(c.hp <= 0 for c in self.enemies) or all(c.hp <= 0 for c in self.allies)):
        # Display the player's stats
            print("\n| Player Name:", self.player.name, "| HP:", self.player.hp, "/", self.player.max_hp, "| Mana:", self.player.mana, "/", 100, "|")

            # Display the allies' stats
            print("\nAllies:")
            for ally in self.allies:
                print(" - ", ally.name, "HP:", ally.hp, "/", ally.max_hp)

            # Display the actions available to the player
            print("\nActions:")
            print("| 1. Attack | 2. Recharge (restores 30 mana) | 3. Fire Bolt |")
            print("| 4. Heal (costs 20 mana) | 5. Mass Heal (costs 30 mana) | 6. Fire Storm (costs 50 mana) |")
            print("| 7. Exit the Game")
            # Ask for input from the user
            action = input("\nEnter the number of the action you would like to take: \n")

            # Check if the input is correct
            while action not in ["1", "2", "3", "4", "5", "6", "7"]:
                print("Invalid input. Try again.")
                action = input("Enter choice: ")

            # Attack
            if action == "1":
                # Print a list of the player's enemies
                print("Which enemy would you like to attack?")
                for i, enemy in enumerate(self.enemies):
                    enemy.hp = max(0, enemy.hp)
                    print("{}. {} HP: {}/{}".format(i + 1, enemy.name, enemy.hp, enemy.max_hp))

                # Get the player's chosen enemy
                enemy_choice = int(input("Enter the number of the enemy you would like to attack: "))
                enemy = self.enemies[enemy_choice - 1]

                # Attack the chosen enemy
                success = self.player.attack(enemy)
                if success:
                    # End the player's turn if the attack was successful
                    return

            # Recharge
            elif action == "2":
                # Recharge the player's mana
                self.player.recharge()

                # Fire bolt
            elif action == "3":
                    # Print a list of the player's enemies
                print("Which enemy would you like to attack with a fire bolt?")
                for i, enemy in enumerate(self.enemies):
                    enemy.hp = max(0, enemy.hp)
                    print("{}. {} HP: {}/{}".format(i + 1, enemy.name, enemy.hp, enemy.max_hp))

                    # Get the player's chosen enemy
                enemy_choice = int(input("Enter the number of the enemy you would like to attack with a fire bolt: "))
                enemy = self.enemies[enemy_choice - 1]

                    # Attack the chosen enemy with a fire bolt
                success = self.player.fire_bolt(enemy)
                if success:
                        # End the player's turn if the fire bolt was successful
                    return

                # Heal
            elif action == "4":
                    # Print a list of the player's allies
                print("Which ally would you like to heal?")
                for i, ally in enumerate(self.allies):
                    print("{}. {} HP: {}/{}".format(i + 1, ally.name, ally.hp, ally.max_hp))

                    # Get the player's chosen ally
                ally_choice = int(input("Enter the number of the ally you would like to heal: "))
                ally = self.allies[ally_choice - 1]

                    # heal the chosen ally
                success = self.player.heal(ally)
                if success:
                        # End the player's turn if the heal was successful
                    return

                # Mass heal
            elif action == "5":
                success = self.player.mass_heal(self.allies)
                if success:
                        # End the player's turn if the mass heal was successful
                    return

            elif action == "6":
                success = self.player.fire_storm(self.enemies)
                if success:
                        # End the player's turn if the fire storm was successful
                    return

            elif action == "7":
                print("You have exited the game! Goodbye")
                exit()

# Create a Battle instance
battle = Battle()

# Start the battle
battle.start()
```
